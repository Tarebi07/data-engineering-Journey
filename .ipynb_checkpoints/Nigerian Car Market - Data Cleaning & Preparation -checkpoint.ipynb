## Understanding the Data
Before cleaning anything, we first understand what we're working with.

```python
def explore_df(df):
    print(df.head())
    print(df.shape)
    print(df.columns)
    print(df.dtypes)

explore_df(df_cars)
```

## Identifying Missing Values
Real world data is messy. Here we find exactly where the gaps are.

```python
print(df_cars.isnull().sum())
```

```python
print(df_cars.shape[0])
```

## Dropping Unusable Columns
Columns missing more than 50% of data are beyond saving.

```python
cols_to_drop =["Trim","Drivetrain","Seats", "Number of Cylinders","Horse Power","Registered city"]
df_cars = df_cars.drop(columns=cols_to_drop)
```

```python
print(df_cars.shape)
print(df_cars.isnull().sum())
```

```python
print(df_cars.shape)
print(df_cars.isnull().sum())
```

## Fixing Remaining Missing Values
Strategic decisions on how to handle each remaining column.

```python
cols_to_drop =["car"]
df_cars = df_cars.drop(columns=cols_to_drop)
```

```python
df_cars =df_cars.dropna(subset=["fuel type", "gear type"])

df_cars["Colour"] = df_cars["Colour"].fillna("Unknown")

df_cars["Mileage"] = df_cars["Mileage"].fillna(df_cars["Mileage"].median())

df_cars["Engine Size"] = df_cars["Engine Size"].fillna(df_cars["Engine Size"].median())
```

```python
print(df_cars.isnull().sum())
```

```python
df_cars.to_csv("car_prices_cleaned.csv", index=False)
print("Clean dataset saved")
```

## Key Principles Applied
- Never fill missing data blindly — ask why it's missing first
- Use median not mean when outliers exist
- Raw data is sacred — always clean into a new file
